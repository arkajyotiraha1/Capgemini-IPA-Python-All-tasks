### Count words in a file

```python
with open('a.txt', 'r') as f:
    words = f.read().split()
    print("Total words:", len(words))
```

### Count words separated by a ','

```python
with open('a.txt', 'r') as f:
    words = f.read().split(',')
    print("Total words:", len(words))
```

### Extract unique words and count them

```python
with open('a.txt', 'r') as f:
    words = f.read().split()
    unique_words = set(words)
    print("Unique words:", unique_words)
```

### Create a dictionary of word count

```python
# Create a dictionary of word count
with open('a.txt', 'r') as f:
    words = f.read().split()
    word_count = {}
    for word in words:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
    print("Word count:", word_count)
```

```python
import json
data = {
    "name" : "arka",
    "age" : 22,
    "course" : "IT"
}
with open("data.json", "w") as file:
    json.dump(data, file)
```
