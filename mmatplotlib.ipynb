```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(1,50,10)
y = np.random.randint(1,30,10)
```

```python
plt.plot(x, y, color = 'red', linestyle = 'dotted')
```

```python
plt.title("Random chart")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.plot(x, y, color = 'red', linestyle = 'dotted')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'green', linestyle = 'dotted')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'chartreuse', linestyle = '-.')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'chartreuse', linestyle = '-.', alpha = 1)
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'teal', linestyle = '-.', alpha = 1, marker = "*")
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'teal', linestyle = '-.', alpha = 1, marker = "*", markerfacecolor = 'y')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'teal', linestyle = '-.', alpha = 1, marker = "*", markerfacecolor = 'y', markersize = 10)
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'teal', linestyle = '-.', alpha = 1, marker = "*", markerfacecolor = 'y', markersize = 10, markeredgecolor = 'red')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'teal', linestyle = '-.', alpha = 1, marker = "*", markerfacecolor = 'y', markersize = 10, markeredgecolor = 'red', linewidth = 2)
plt.show()
```

```python
x = np.arange(1, 11)
y = np.sin(x)
z = np.cos(x)
```

```python
plt.plot(x, y, label = 'SINx', color = 'teal')
plt.plot(x, z, label = 'COSx', color = 'pink')
plt.title("Sinx and Cosx Graph")
plt.xlabel("Data Points")
plt.ylabel("Associate values")
plt.xticks()
plt.yticks()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(x, y, label = 'SINx', color = 'teal')
plt.plot(x, z, label = 'COSx', color = 'pink')
plt.title("Sinx and Cosx Graph")
plt.xlabel("Data Points")
plt.ylabel("Associate values")
plt.xticks()
plt.yticks()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,9))
plt.grid()

plt.subplot(3, 3, 1)
plt.title("SINQ and COSQ graph")
plt.plot(x,y,label='sinq',color='blue')
plt.plot(x,z,label='cosq',color='red')
plt.xlabel("x value")
plt.ylabel("Associated value")
plt.xticks()
plt.yticks()
plt.grid()

plt.legend(loc = "upper right")

plt.subplot(3,3,5)
plt.title("SINQ and COSQ2 graph")
plt.plot(x,y,label='sinq',color='orange')
plt.plot(x,z,label='cosq',color='teal')
plt.xlabel("x value")
plt.ylabel("Associated value")
plt.xticks()
plt.yticks()
plt.grid()
plt.legend(loc = "upper right")

plt.subplot(3,3,9)
plt.title("SINQ and COSQ2 graph")
plt.plot(x,y,label='sinq',color='yellow')
plt.plot(x,z,label='cosq',color='cyan')
plt.xlabel("x value")
plt.ylabel("Associated value")
plt.xticks()
plt.yticks()
plt.grid()
plt.legend(loc = "upper right")


plt.tight_layout()
plt.show()
```
