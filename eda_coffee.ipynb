```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("coffee.csv")
```

```python
print("1) Rows and Columns:")
print(df.shape)
```

```python
print("2) Column Names:")
print(df.columns)
```

```python
print("3) Data Types:")
print(df.dtypes)
```

```python
print("4) First 5 Rows:")
print(df.head())
```

```python
print("5) Last 5 Rows:")
print(df.tail())
```

```python
print("6) Missing Values:")
print(df.isnull().sum())
```

```python
print("7) Unique Countries:")
print(df['Country'].nunique())
```

```python
print("8) Records per Year:")
print(df['Year'].value_counts())
```

```python
print("9) Records per Country:")
print(df['Country'].value_counts())
```

```python
print("10) Production Statistics:")
print("Minimum:", df['Total Production'].min())
print("Maximum:", df['Total Production'].max())
print("Average:", df['Total Production'].mean())
```

```python
print("11) Country with Highest Single Record Production:")
max_row = df.loc[df['Total Production'].idxmax()]
print(max_row['Country'], "-", max_row['Total Production'])
```

```python
print("\n12) Missing or Zero Production Rows:")
print(df[(df['Total Production'].isnull()) | (df['Total Production'] == 0)])
```

```python
print("13) Average Import Price:")
print(df['Import Price'].mean())
```

```python
print("14) Total Production Per Year:")
production_year = df.groupby('Year')['Total Production'].sum()
print(production_year)
```

```python
print("15) Top 5 Countries by Production:")
top5 = df.groupby('Country')['Total Production'].sum().sort_values(ascending=False).head()
print(top5)
```

```python
print("16) Highest Average Export Price:")
avg_export = df.groupby('Country')['Export Price'].mean()
print(avg_export.idxmax(), "-", avg_export.max())
```

```python
print("17) Domestic Consumption greater than Production:")
print(df[df['Domestic Consumption'] > df['Total Production']])
```

```python
print("18) Highest Growth Country:")
growth = df.groupby('Country')['Total Production'].apply(lambda x: x.max() - x.min())
print(growth.idxmax(), "-", growth.max())
```

```python
print("19) Correlation between Export Price and Production:")
print(df['Export Price'].corr(df['Total Production']))
```

```python
top5.plot(kind='bar')
plt.title("Top 5 Coffee Producing Countries")
plt.xlabel("Country")
plt.ylabel("Total Production")
plt.show()


production_year.plot()
plt.title("Total Coffee Production per Year")
plt.xlabel("Year")
plt.ylabel("Total Production")
plt.show()

plt.scatter(df['Total Production'], df['Export Price'])
plt.title("Export Price vs Total Production")
plt.xlabel("Total Production")
plt.ylabel("Export Price")
plt.show()
```
