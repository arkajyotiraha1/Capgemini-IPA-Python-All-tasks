```python
from abc import ABC, abstractmethod
class BankAccount(ABC):
    def __init__(self, customer_name, date_of_birth, balance):
        self.__customer_name = customer_name
        self.__date_of_birth = date_of_birth
        self._balance = balance
    def get_customer_name(self):
        return self.__customer_name
    def get_date_of_birth(self):
        return self.__date_of_birth
    def get_balance(self):
        return self._balance
    @abstractmethod
    def deposit(self, amount):
        pass
    @abstractmethod
    def withdraw(self, amount):
        pass
class SavingsAccount(BankAccount):
    def __init__(self, customer_name, date_of_birth, balance):
        super().__init__(customer_name, date_of_birth, balance)
        self.__minimum_balance = 1000
    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print("Deposited:", amount)
        else:
            print("Invalid deposit amount")
    def withdraw(self, amount):
        if amount > 0 and (self._balance - amount) >= self.__minimum_balance:
            self._balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Withdrawal failed! Minimum balance must be 1000.")
if __name__ == "__main__":
    account = SavingsAccount("Arka", "15-08-2003", 10000)
    account.deposit(3000)
    account.withdraw(5000)
    account.withdraw(7000)
    account.withdraw(1000)
    print("Final Balance:", account.get_balance())


   
```
